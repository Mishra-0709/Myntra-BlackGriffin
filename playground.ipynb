```python
import json
import matplotlib.pyplot as plt
import numpy as np
import stone
from PIL import Image
import random
```

```python
with open('data.json', 'r') as f:
    json_data = json.load(f)
```

```python
image_path = 'mike.jpg'
image = Image.open(image_path)
res = stone.process(image_path,'color')
skin_tone = res['faces'][0]['skin_tone']
#get three random color recommendations from data.json
plt.imshow(image)
suggestions = random.sample(json_data[skin_tone],3)
fig, axs = plt.subplots(1, 3, figsize=(5, 3))

for i, color in enumerate(suggestions):
    r, g, b = color
    axs[i].imshow([[[r/255, g/255, b/255]]])
    axs[i].axis('off')
```
